#### Subplot y subplots

Vimos algunas opciones para modificar y personalizar los gráficos. Ahora volvamos sobre como hacer varios gráficos juntos en una misma figura. Para ello podemos usar los métodos subplot() y subplots(), noten que hay una "s" que las diferencia. Primero veamos subplot().

```python
plt.figure()

plt.subplot(311)
plt.plot(x, y, 'y--+', label = 'y')

plt.subplot(312)
plt.plot(x, y2, 'k-.D', label = 'y2')

plt.subplot(313)
plt.plot(x, y3, ':', label = 'y3')

plt.show()
```

```python
plt.figure()

plt.subplot(131)
plt.plot(x, y, 'y--+', label = 'y')

plt.subplot(132)
plt.plot(x, y2, 'k-.D', label = 'y2')

plt.subplot(133)
plt.plot(x, y3, ':', label = 'y3')

plt.show()
```

```python
plt.figure()

plt.subplot(131)
plt.plot(x, y, 'y--+', label = 'y')

plt.subplot(132)
plt.plot(x, y2, 'k-.D', label = 'y2')

plt.subplot(133)
plt.plot(x, y3, ':', label = 'y3')
plt.xlabel('Variable independiente', fontsize = 12)
plt.ylabel('Variable dependiente', fontsize = 12)
plt.title('Gráfico N°3')
plt.legend(title = 'Líneas')

plt.suptitle('Subplot', fontsize = 14)
plt.tight_layout()
plt.show()
```

Ahora estudiemos un poco subplots().

```python
fig, ax = plt.subplots(1, 3)

ax[0].plot(x, y, 'y--+', label = 'y')
ax[1].plot(x, y2, 'k-.D', label = 'y2')
ax[2].plot(x, y3, ':', label = 'y3')

plt.tight_layout()
plt.show()
```

Tambien podemos modificar cada eje por separado, sin embargo los métodos cambian un poco:

<center><img src = "https://matplotlib.org/stable/_images/anatomy.png" width = 65% high = 65%><center>

```python
fig, ax = plt.subplots(1, 3)

ax[0].plot(x, y, 'y--+', label = 'y')
ax[1].plot(x, y2, 'k-.D', label = 'y2')
ax[2].plot(x, y3, ':', label = 'y3')

for i in range(3):
  ax[i].set_xlabel('Variable independiente', fontsize = 12)
  ax[i].set_ylabel('Variable dependiente', fontsize = 12)
  ax[i].legend(title = 'Líneas')

plt.tight_layout()
plt.show()
```

Podemos hacer que los gráficos queden "pegados" por un eje usando sharex y/o sharey junto al método subplots_adjust. Esto es de utilidad si estás graficando la misma variable en el eje a unir.

```python
fig, ax = plt.subplots(1, 3, figsize = (10, 4.8), sharey = True)
fig.subplots_adjust(wspace=0)

ax[0].plot(x, y, 'y--+', label = 'y')
ax[1].plot(x, y2, 'k-.D', label = 'y2')
ax[2].plot(x, y3, ':', label = 'y3')

ax[0].set_ylabel('Variable dependiente', fontsize = 12)
ax[1].set_xlabel('Variable independiente', fontsize = 12)

for i in range(3):
  ax[i].legend(title = 'Líneas')

plt.show()
```

Podemos agregar todos los gráficos que querramos a un subplots, tenemos que explicitar en que posición va cada gráfico. Veamos dos maneras de hacerlo:

```python
y4 = [i**2 + 6 for i in range(0, 6)]

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

ax1.plot(x, y, 'y--+', label = 'y')
ax2.plot(x, y2, 'k-.D', label = 'y2')
ax3.plot(x, y3, ':', label = 'y3')
ax4.plot(x, y4, 'go', label = 'y4')

fig.text(0.5, -0.025, 'Variable independiente', fontsize = 12, ha = 'center')
fig.text(-0.025, 0.5, 'Variable dependiente', fontsize = 12 , va = 'center', rotation = 'vertical')

plt.suptitle('Grilla 2x2')

fig.legend(bbox_to_anchor = [1.1, 0.915])

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)

ax[0,0].plot(x, y, 'y--+', label = 'y')
ax[0,1].plot(x, y2, 'k-.D', label = 'y2')
ax[1,0].plot(x, y3, ':', label = 'y3')
ax[1,1].plot(x, y4, 'go', label = 'y4')

fig.text(0.5, -0.025, 'Variable independiente', fontsize = 12, ha = 'center')
fig.text(-0.025, 0.5, 'Variable dependiente', fontsize = 12 , va = 'center', rotation = 'vertical')

plt.suptitle('Otra grilla 2x2')

fig.legend(bbox_to_anchor = [1.1, 0.915])

plt.tight_layout()
plt.show()
```

### Tipos de gráficos

Ahora veamos algunos gráficos, pensando siempre cual es su utilidad, para que tipo de datos sirven y cual es el método de Matplotlib correspondiente.

Antes que nada, carguemos un par de datasets más interesantes para hacer uso de los gráficos.

```python
reservas_path = 'https://raw.githubusercontent.com/MauriJ/Visualizacion_Bancor/main/Datasets/reservas-internacionales-pasivos-financieros-bcra.csv'
df_reservas = pd.read_csv(reservas_path)
df_reservas['indice_tiempo'] = pd.to_datetime(df_reservas['indice_tiempo'])
df_reservas.head()
```

```python
credito_path = 'https://raw.githubusercontent.com/MauriJ/Visualizacion_Bancor/main/Datasets/credit_risk_dataset.csv'
df_credito = pd.read_csv(credito_path)
df_credito.head()
```

#### Gráfico de línea



Los gráficos de líneas son útiles para revelar tendencias o cambios a lo largo del tiempo. También son utilizados para mostrar relaciones dentro de un conjunto de datos contínuo.

```python
plt.plot(df_reservas['indice_tiempo'], df_reservas['reservas_internacionales_dolares'], lw = 3)

plt.xlabel('Año')
plt.ylabel('Reservas (en millones de dólares)')

plt.show()
```

#### Gráfico de puntos o de dispersión

Estos gráficos sirven para representar visualmente la relación entre dos variables, útiles para indentificar patrones, tendencias, correlaciones, valores atípicos y un largo etcétera.

```python
plt.scatter(df_credito['person_income'], df_credito['loan_amnt'], s = 3)
plt.xscale('log')

plt.xlabel('Ingreso')
plt.ylabel('Monto del préstamo')

plt.show()
```

```python
plt.scatter(df_credito['person_income'].sample(100), df_credito['loan_amnt'].sample(100),
         color = 'C1',
         edgecolors = 'black',
         linewidth = 1,
         s = 10)

plt.xscale('log')

plt.xlabel('Ingreso')
plt.ylabel('Monto del préstamo')

plt.show()
```

#### Gráfico de barras

Los gráficos de barras son ideales para comparar datos entre diferentes categorías o grupos. Cada barra representa una categoría o grupo, y la altura de la barra refleja la magnitud o valor de la variable que se está representando. También se utilizan para mostrar la frecuencia de aparición de categorías o valores discretos en un conjunto de datos.

```python
ejex = df_credito['person_home_ownership'].value_counts().index
ejey = df_credito['person_home_ownership'].value_counts().values

plt.bar(x = ejex, height = ejey)

plt.xlabel('Titularidad')
plt.ylabel('Conteo')

plt.show()
```

También podemos graficarlos en formato horizontal.

```python
ejey = df_credito['person_home_ownership'].value_counts(ascending = True).index
ejex = df_credito['person_home_ownership'].value_counts(ascending = True).values

plt.barh(y = ejey, width = ejex, color = 'yellow', edgecolor = 'black', hatch = '/')

plt.ylabel('Titularidad')
plt.xlabel('Conteo')

plt.show()
```

#### Gráficos de torta

Los gráficos de torta, frecuentemente vilipendiados, tiene utilidad similar a los gráficos de barras con respecto a comparar variables categóricas. Son útiles para representar proporciones.

```python
labels = df_credito['person_home_ownership'].value_counts().index
x = df_credito['person_home_ownership'].value_counts().values

plt.pie(x = x, labels = labels, startangle = 90, autopct = '%1.1f%%')
plt.show()
```

#### Histogramas

Los histogramas son extremadamente útiles para analizar la distribución de valores en una variable numérica contínua. Los van a usar muy frecuentemente!

```python
plt.hist(df_credito['loan_amnt'], bins = 20, edgecolor='black')

plt.xlabel('Monto del préstamo')
plt.ylabel('Frecuencia')

plt.show()
```

Si pasamos el parámetro density = True obtenemos la densidad de probabilidad. Esto "normaliza" el gráfico, haciendo que el área debajo del histograma sea igual a 1.

```python
plt.hist(df_credito['loan_amnt'], bins = 20, density = True, color = 'C3', edgecolor='black', lw = 2)

plt.xlabel('Monto del préstamo')
plt.ylabel('Densidad de probabilidad')

plt.show()
```

#### Gráficos de cajas y bigotes

Los gráficos de cajas, también conocidos como diagramas de caja o box plots en inglés, condensan una gran cantidad de información estadística que permiete dar una idea de la distribución de valores en una variable. Visualizan la mediana, los cuartiles Q1 y Q3, el mínimo y máximo inlier y los outliers.

<center><img src = "https://www.qvision.es/blogs/wp-content/uploads/2021/03/457px-Boxplot.svg.png" width = 50% high = 50%><center>

```python
plt.boxplot([df_credito['loan_amnt'], df_credito['person_income']], labels = ['Monto del préstamo', 'Ingresos'])

plt.ylim(None, 2000000)

plt.xlabel('Variable')
plt.ylabel('Valores de las variables')

plt.show()
```

Si explicitamos el parámetro showfliers = False hacemos que no se muestren los outliers. Repito, no los borra, simplemente no los grafica.

```python
plt.boxplot([df_credito['loan_amnt'], df_credito['person_income']], labels = ['Monto del préstamo', 'Ingresos'], showfliers = False)

plt.xlabel('Variable')
plt.ylabel('Valores de las variables')

plt.show()
```

Y el parámetro vert = False nos permite graficar las cajas horizontalmente.

```python
bplot = plt.boxplot([df_credito['loan_amnt'], df_credito['person_income']],
            labels = ['Monto del préstamo', 'Ingresos'],
            vert = False,
            notch=True,
            showfliers = False,
            patch_artist = True)

colors = ['lightblue', 'lightgreen']

for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

plt.ylabel('Variable')
plt.xlabel('Valores de las variables')

plt.show()
```

#### Gráfico de violín

Los gráficos de violín parten de la misma idea de los gráficos de caja pero representan, en cambio, la densidad de probabilidad. Estos gráficos también son utilizados para ilustrar la distribución de las variables.

```python
plt.violinplot([df_reservas['pasivos_monetarios_letras_pesos'], df_reservas['pasivos_monetarios_letras_dolares']])

plt.xticks([1, 2], ['Pesos', 'Dólares'])

plt.xlabel('Pasivos monetarios en letras')
plt.ylabel('Monto')

plt.show()
```
